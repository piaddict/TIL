# 랭크, 역행렬, 일차방정식

## 문제설정: 역문제

지금까지와 반대로 결과를 알고있는 상태에서 원인을 추측하고 싶은 경우도 있다.
이처럼 결과 y를 먼저 알고 원인 x를 추정하는 형태의 문제를 역문제라고 한다.

책에서는 문제상황의 노이즈(constant가 존재하는 경우)를 배제하고 우선 문제해결을 한다.

## 성질이 좋은 경우 (정칙행렬)

### 정칙성과 역행렬

역행렬이 존재하는 정방행렬 A를 정칙행렬이라고 한다.
정칙이 아닌 행렬은 특이행렬이라고 한다.

### 기본변형

1. 어느 행을 c배 한다 ($c \neq 0$)
2. 어느 행의 c배를 다른행에 더한다
3. 어느 행과 다른 행을 바꿔넣는다

### 연립일차방적식의 해법 (정칙인 경우)

구체적으로 주어진 $A$와 $y$에 대해 $Ax = y$가 되는 $x$를 구하는 이야기이다.

평소대로 연립 일차방정식을 풀어서 해결할 수 있다 (기본변형 활용)

1. 변수소거로 연립방정식 풀기: 변수 하나의 값을 우선구한 뒤 차례로 나머지 변수 값을 구함
2. 가우스 요르단 소거법: 블록행렬 표기만으로 연립일차방정식 풀기 -> 기약 행사다리꼴로 변환 -> 단위행렬로 변환

### 역행렬의 계산

1. 연립일차방정식의 해를 구하여 대입하면 구할 수 있다
2. 블록행렬 표기로 푼다 ($A|I$ -> $A|A^{-1}$)
