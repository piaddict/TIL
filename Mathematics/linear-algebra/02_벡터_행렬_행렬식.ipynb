## 행렬과 사상

### 순수한 관계를 나타냄

수를 행과 열의 사각형 형태로 나열한 것을 행렬이라고 함, 특히 행과 열 수가 같은 경우 정방행렬이라 부른다.


### 행렬은 사상이다

n차원 벡터 x에 $m \times n$ 행렬 A를 곱하면 m차원 벡터 $y = Ax$를 얻을 수 있다

행렬 A를 지정하면 벡터를 다른 벡터에 옮기는 Mapping이 결정된다.

행렬 $A = \begin{bmatrix} 1 & -0.3 \\ -0.7 & 0.6 \end{bmatrix}$

$e_{1} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$ 를 $\begin{bmatrix} 1 \\ -0.7 \end{bmatrix}$ 에

$e_{2} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$ 를 $\begin{bmatrix} -0.3 \\ 0.6 \end{bmatrix}$ 으로 mapping

사상이 같으면 행렬도 같으므로 행렬 A, B가 임의의 벡터 x에 대해 항상 Ax = Bx 라면 A = B가 성립

### 행렬의 곱 = 사상의 합성

행렬은 열(column)을 기준으로 가로로 확장함

$k \times n$ 행렬과 $m \times n$ 행렬의 곱은 $k \times n$ 행렬임

벡터 x를 A에 mapping 하여 옮긴 결과 y = Ax 를 다시 B로 mapping한 결과인 z = B(Ax)를 행렬 B에 A를 곱한것으로 볼 수 있다. (순서 A -> B)

### 행렬 연산의 성질

- (cA)x = c(Ax) = A(cx)
- (A + B)x = Ax + Bx
- A + B = B + A
- (A + B) + C = A + (B + C)
- (c + c')A = cA + c'A
- (cc')A = c(c'A)
- A(B + C) = AB + AC
- (A + B)C = AC + BC
- (cA)B = c(AB) = A(cB)

#### 벡터와 행렬

n차원 벡터를 $n \times 1$ 행렬로 간주하여 연산하여도 결과는 같다.

횡 곱하기 종과 종 곱하기 횡은 결과가 다르므로 주의해야 한다. (기본적으로 종 벡터를 다룸)

#### 1 X 1 행렬과 수

기저를 변환하였을 때 성분이 어떻게 변하는가.

수는 기저를 변환하여도 값이 변하지 않지만, 1차원 벡터$\vec{v} = v_{1}\vec{e}_{1}$는 기저($\vec{e}_{1}$를 바꾸면 성분($v_{1}$)도 변한다.

또한 1X1 행렬에도 여러 종류가 있으므로 일률적으로 말할 수 없다. (쌍대공간 참고)

### 영행렬, 단위행렬, 대각행렬

#### 영행렬

모든 성분이 0인 행렬을 **영행렬**이라 하고 기호 $O$로 나타냄

영행렬의 사상은 모든 것을 원점으로 이동시키는 것

임의의 행렬 A에 대하여 영행렬과의 연산은 다음과 같다

- $A + O = O + A = A$
- $AO = O$
- $OA = O$
- $0A = O$

또한 $ A \neq O, B \neq 0$ 인데도 $BA=O$가 가능하다.
마찬가지로 $A \neq 0$인데도 $A^2 = O$가 가능하다.

#### 단위행렬

정방행렬에서 대각선의 요소가 모두 1이고 나머지는 모두 0인 행렬을 단위행렬이라고한다.

단위행렬은 기호 $I$로 나타낸다.

단위 행렬의 사상은 아무것도 하지 않는(공간 이동이 없는) 사상이다.

#### 대각행렬

정방행렬의 대각선상의 값을 대각성분이라고 한다.

비대각 성분이 모두 0인 행렬을 대각행렬이라고 부른다.

대각행렬은 다음과 같이 줄여쓴다. $diag(a_{1}, a_{2} \dots a_{n})$

대각행렬인 나타내는 사상은 축에 따라 늘고 주는 것이고 대각 성분의 크기/여하에 따라 공간이 변하는 모습도 달라진다.

대각행렬에서는 $y = Ax$가 독립된 n개로 분리되어 있다

$y_{1} = a_{1}x_{1} \dots y_{n} = a_{n}x_{n}$

### 역행렬 = 역사상

A에 이동시킨 것을 원래대로 돌려놓는다는 것과 같음

#### 정의

정방행렬 $A$에 대해 그 역 사상에 대응하는 행렬을 A의 역행렬이라고 하고 $A^{-1}$이라고 쓴다

어떠한 $x$를 가져와도 $Ax = y$ 또는 $A^{-1} y = x$이고 반대로 어떠한 $y$에도 $A^{-1} y = x$ 또는 $Ax = y$가 되는 행렬 $A^{-1}$이다.

다음을 표현하면 $A^{-1} A = AA^{-1} = I$

역행렬은 있을 수도, 없을 수도 있다. 또한 역행렬은 1개 존재하고, 정방행렬에 대해서만 존재한다.

#### 기본적인 성질

- $(A^{-1})^{-1} = A$
- $(AB)^{-1} = B^{-1} A^{-1}$
- $(A^k)^{-1} = (A^{-1})^k$
- $A^{-1}$에 $A$를 곱하면 $I$이므로 $A$는 $A^{-1}$의 역행렬
- $(AB)(B^{-1}A^{-1}) = ABB^{-1} A^{-1} = A(BB^{-1})A^{-1} = AIA^{-1} = AA^{-1} = I$
- $ A^{k}(A^{-1})^{k} = a^{k-1}AA^{-1}(A^{-1})^{k-1} = A^{k-1}I(A^{-1})^{k-1} = A^{k-1}(A^{-1})^{k-1} = \dots = AA^{-1} = I$

개수가 늘어도 다음과 같다 $(ABCD)^{-1} = D^{-1}C^{-1}B^{-1}A^{-1}$

#### 대각행렬의 경우

대각행렬의 원소 $a_{1} \dots a_{n}$중 하나라도 0이 있다면 역행렬을 만들 수 없다.

### 블록행렬

#### 정의와 성질

행렬의 종횡에 단락을 넣어 각 구역을 작은 행렬로 간주한 것을 블록행렬이라 한다.

- 블록행렬의 덧셈
- 블록행렬의 정수배
- 블록행렬의 곱

블록 행렬의 곱에서 블록행렬은 마치 $B_{ij}$와 $A_{ij}$가 보통의 수인 것 처럼 행렬곱 모양으로 계산해도 상관없으나, 순서에는 주의해야 한다.

#### 행벡터, 열벡터

블록행렬의 특별한 경우로 한 방향으로만 작게 나누는 것도 생각할 수 있다.

각 단편의 크기가 nx1(행벡터)이나 1xn(열벡터)인 경우를 생각해보자.

#### 블록대각행렬

대각선상의 블록이 모두 정방행렬이고, 그외의 블록이 모두 영행렬인 것을 블록 대각행렬이라고 한다.
