### 행렬식과 확대율

행렬의 면적 확대율에 관한 것을 그 행렬의 행렬식(determinant)라고 한다.
3차 정방행렬의 경우라면 행렬식은 부피 확대율을 말하게 된다.
이를 일반적으로 말하면, n차 정방행렬 A에 대해 n차원의 부피 확대율이 `det A`이다.

`det A` 또는 `|A|` 같은 방식으로 표기한다.

> 적분은 그래프의 면적, 이중적분은 3차원 그래프의 부피이다. 따라서 다중적분의 변수 변환에서는 행렬식이 본질적인 역할을 하게 된다. (Jacobian참고)

> 밀도에서도 행렬식을 사용할 수 있다. 공간이 늘어나면 밀도가 내려가기 때문이다.

> 행렬식은 또한 역행렬의 존재여부를 판단할 수 있게 한다. 축 하나가 납작하게 눌리는 경우(다른 점들이 같은 점으로 이동하는) 다시 공간을 돌릴 방법이 없다.

### 행렬식의 성질

- $det I = I$
- $det(AB) = (detA)(detB)$
- $detA^{-1} = \dfrac{1}{detA}$
- $detA = 0$ 이면 $A^{-1}$은 존재하지 않는다 <- $(detA)(detA^{-1}) = 1$ 이 성립해야 한다
- $det(diag(a_{1}, \dots, a_{n})) = a_{1} \dots a_{n}$
- 하나 이상의 열이 모두 0인 경우나 어느 두 열 이상이 완전히 같은경우 $det A = 0$

#### 유용한 성질

행렬식은 어느 열의 정수배를 다른 열에 더해도 값이 변하지 않는다

> 축의 방향이 변해도 전체 면적(부피)가 변하지 않는다는 것을 떠올려 보자 (책에서는 쌓아놓은 트럼프를 미는 예제로 설명함)

행렬이 특별한 모양이면 행렬식을 간단히 구할 수 있다.

대각성분 보다 아래쪽이 모두 0인 행렬을 상삼각행렬이라고 한다.(위쪽이 모두 0인 경우 하삼각행렬) 이 경우 행렬식은 대각행렬의 경우와 같다.

#### 전치행렬의 행렬식

전치행렬의 행렬식은 원래 행렬의 행렬식과 같다: $det(A^{T}) = detA$

- 어느 행의 정수배를 다른 행에 더해도 행렬식은 동일함
- 하삼각행렬의 행렬식은 대각성분의 곱이다

#### 중요한 성질: 다중선형성, 교대성

행렬식이 n차원의 부피라면 다중선형성도 도형으로 이해할 수 있다

- $det(ca_{1}, a_{2}, \dots, a_{n}) = cdet(a_{1}, a_{2}, \dots, a_{n})$
- $det(a_{1} + a'_{1}, a_{2}, \dots, a_{n}) = det(a_{1}, a_{2}, \dots, a_{n}) + det(a'_{1}, a_{2}, \dots, a_{n})$

위의 식을 해석해 본다면

- (1)의 경우: 특정 방향의 벡터만 c배 되면 부피도 c배 됨
- (2)의 경우: 트럼프를 약간 민 것으로 해석 가능

결국 다중선형성은 한 열만 조작하는 경우를 두고 이야기 하는 것

만약 n차 정방행렬 A를 c배 하면 행렬식은 $c^{n}$배가 된다: $det(cA) = c^{n}detA$

교대성: 행렬에서 두 열을 바꾸면(도형을 뒤집음: 거울상) 행렬식의 부호가 역전된다

$det(a_{2}, a_{1}, a_{3}) = -det(a_{1}, a_{2}, a_{3})$


### 행렬식 계산법: 수식 계산

$detA = \sum_{i_{1},\dots,i_{n}} \epsilon_{i_{1} \dots i_{n}}a_{i_{1}1} \dots a_{i_{n}n}$

$\epsilon_{i_{1} \dots i_{n}}$의 경우, 앞의 성질과 동일하다

- $\epsilon_{123} = 1$
- 첨자가 교체되면 부호가 변함: $\epsilon_{123} = -\epsilon_{213} = 1$
- 첨자에 같은것이 있는 경우 0: $\epsilon_{113} = 0$
