### 좌표 변환과 행렬

<http://mrw0119.tistory.com/m/12>

같은 공간에서 기저를 취할 수 있는 방법은 여러가지 이지만 어떻게 기저를 취하여도 벡터 실체는 같음

- 좌표 변환은 정방행렬 A를 곱한다는 형태로 쓸 수 있다. A에는 역행렬이 존재한다
- 역행렬을 지니는 정방행렬 A를 곱하는 것은 좌표 변환이라고 해석할 수 있음

좌표 변환과 역변환 (정방행렬 A와 그 역행렬 A')

정방행렬을 곱할 때 역행렬이 존재한다면 좌표변환이라고 해석할 수 있다.

$x_{1}\vec{e}_{1} + \dots + x_{n}\vec{e}_{n} = x'_{1}\vec{e'}_{1} + \dots + x'_{n}\vec{e'}_{n}$

위의 식은 어느 벡터를 기저$(\vec{e'}_{1}, \dots, \vec{e'}_{n})$로 표현할 때의 좌표가 $(x'_{1}, \dots, x'_{n})^{T}$이고,
다른 기저 $(\vec{e}_{1}, \dots, \vec{e}_{n})$로 표현할 때의 좌표가 $(x_{1}, \dots, x_{n})^{T}$라는 형태이다.

하지만 $(\vec{e'}_{1}, \dots, \vec{e'}_{n})$가 기저의 조건을 만족하고 있는지 알 수 없으므로 A의 역행렬을 구하여 증명한다.

$(\vec{e}_{1}, \dots, \vec{e}_{n})$는 기저라는 전제이므로 $v' = A^{-1}v$라고 취하면
$x'_{1}\vec{e'}_{1} + \dots + x'_{n}\vec{e'}_{n}$가 성립하는 $v'$는 유일하게 존재한다.

### 전치행렬

행렬 A의 행과 열을 바꿔넣는 것을 A의 전치행렬이라 하고 $A^{T}$라고 쓴다. T는 Transpose이다.

- 전치의 전치는 원래대로 돌아온다.
- 대각행렬 D에 대해 D의 전치행렬은 D이다.
- $(AB)^{T} = B^{T}A^{T}$
- $(A^{-1})^{T} = (A^{T})^{-1}$

사상에서 전치의 의미는 선형공간에서 답하기는 어렵다. (내적과 관련)
전치는 대부분 실 행렬에 대한 이야기임.
